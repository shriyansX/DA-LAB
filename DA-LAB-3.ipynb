```python
num = int(input("Enter a number: "))

temp = num   
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
num = int(input("Enter a number: "))

temp = num
sum_of_powers = 0

digits = len(str(num))

while num > 0:
    digit = num % 10
    sum_of_powers += digit ** digits
    num //= 10

if temp == sum_of_powers:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print("GCD is:", a)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

x, y = a, b
while y != 0:
    x, y = y, x % y
gcd = x

lcm = (a * b) // gcd

print("LCM is:", lcm)
```

```python
for num in range(2, 101):
    is_prime = True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```

```python
start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

count = 0

for num in range(start, end + 1):
    if num > 1:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            count += 1

print("Total prime numbers:", count)
```

```python
num = int(input("Enter a number: "))

even_sum = 0
odd_sum = 0

while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even_sum += digit
    else:
        odd_sum += digit
    num //= 10

print("Sum of even digits:", even_sum)
print("Sum of odd digits:", odd_sum)
```

```python
num = int(input("Enter a number: "))

sum_div = 0

for i in range(1, num):
    if num % i == 0:
        sum_div += i

if sum_div == num:
    print("Perfect number")
else:
    print("Not a perfect number")
```

```python
num = int(input("Enter a number: "))

count = 0

while num > 0:
    digit = num % 10
    if digit % 3 == 0:
        count += 1
    num //= 10

print("Number of digits divisible by 3:", count)
```
